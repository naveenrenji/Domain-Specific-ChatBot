```python
import pandas as pd

# Load the CSV file
file_path = 'section_average.csv'
data = pd.read_csv(file_path)
```

```python
data
```

```python
df = pd.DataFrame(data)
df['Age'] = 2023 - df['Birth_year']
df.to_csv('section_average_updated.csv', index=False)
# data.to_csv(output_file, index=False)
```

```python
df = pd.DataFrame(data)

print(df)

# Change the gender id into labels
df['Gender'] = df['Gender'].replace({1: 'male', 2: 'female', 3: 'non-binary', 4: 'prefer not to say'})
```

```python
# Age category

df_age_cat = df[df['Age'] != 0]

df_age_cat['Age_Category'] = pd.cut(df['Age'], bins=[0, 25, float('inf')], labels=['<25', '>=25'])

df_age_cat
```

```python
## Age plots

import matplotlib.pyplot as plt
import seaborn as sns
from docx import Document
from io import BytesIO
from matplotlib.backends.backend_pdf import PdfPages


with PdfPages('age_scatter_plots.pdf') as pdf:
    # Create scatter plots using seaborn, each in a new line
    for i, column in enumerate(df_age_cat.columns[1:]):
        plt.figure(figsize=(6, 4))  # Adjust the figsize as needed
        sns.scatterplot(x='Age_Category', y=column, data=df_age_cat)
        plt.title(f'Scatter Plot: Age vs {column}')
        plt.xlabel('Age')
        plt.ylabel(column)
#         plt.show()
        
        # Save the scatter plot to the PDF file
        pdf.savefig()

        # Close the current figure to free up resources
        plt.close()      
```

```python
# Create a PDF file to save all box plots
with PdfPages('age_box_plots.pdf') as pdf:
    # Create box plots using seaborn, each in a new line
    for i, column in enumerate(df_age_cat.columns[1:]):
        
        try:
            plt.figure(figsize=(6, 4))  # Adjust the figsize as needed
            sns.boxplot(x='Age_Category', y=column, data=df_age_cat)
            plt.title(f'Box Plot: Age vs {column}')
            plt.xlabel('Age')
            plt.ylabel(column)

            # Save the box plot to the PDF file
            pdf.savefig()

            # Close the current figure to free up resources
            plt.close()
        except:
            print("Error")
```
