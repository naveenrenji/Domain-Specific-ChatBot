```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from docx import Document
from io import BytesIO
from matplotlib.backends.backend_pdf import PdfPages

# Load the CSV file
file_path = 'section_average_updated.csv'
data = pd.read_csv(file_path)
```

```python
df = pd.DataFrame(data)
df.iloc[0]
```

```python
### Gender analysis part 2

male_df = df[df['Gender'] == 1]
male_df['Gender'] = 'Male'

female_df = df[df['Gender'] == 2]
female_df['Gender'] = 'Female'
```

```python
male_df
```

```python
female_df
```

```python
averages = male_df.mean().to_dict()

# Add a new row with averages to the DataFrame
male_df = male_df.append(averages, ignore_index=True)
male_df['Gender'] = 'Male'
```

```python
male_df.iloc[-1]
```

```python
averages = female_df.mean().to_dict()

# Add a new row with averages to the DataFrame
female_df = female_df.append(averages, ignore_index=True)
female_df['Gender'] = 'Female'
```

```python
female_df.iloc[-1]
```

```python
gender_averages = pd.concat([male_df.iloc[-1], female_df.iloc[-1]], axis=1).transpose()
gender_averages = gender_averages.drop(columns=['email'])
```

```python
gender_averages
```

```python
# # Set the index to 'Name' column
# gender_averages.set_index('Gender', inplace=True)

# Initialize a PDF file to save the plots
pdf_pages = PdfPages('gender_averages_histograms.pdf')

# Plot histograms for each column
for column in gender_averages.columns:
    if column != 'Gender' and column !='Age' and column != 'Race' and column != 'Q32_avg':
        gender_averages[column].plot(kind='bar', alpha=0.5)
        plt.xlabel('Gender')
        plt.ylabel(column)
        plt.title(f'Histogram of {column} by Gender')
        max_value = gender_averages[column].max()
        if max_value <= 5:
            plt.ylim(0, 6)
        else:
            plt.ylim(0, 11)
        plt.xticks(range(len(gender_averages['Gender'].unique())), gender_averages['Gender'].unique(), rotation=0)  # Rotate x-axis labels to make them readable

        # Add text labels to each bar
        for i, value in enumerate(gender_averages[column]):
            plt.text(i, value + 0.1, str(value), ha='center', va='bottom')
        plt.show()
        pdf_pages.savefig()  # Save the current figure into the PDF
        plt.close()  # Close the figure to free up memory

# Close the PDF file after saving all plots
pdf_pages.close()
```

```python
df_age_cat = df
df_age_cat['Age_category'] = pd.cut(df['Age'], bins=[0, 25, float('inf')], labels=['<25', '>=25'])
```

```python
lt_25_df = df_age_cat[df_age_cat['Age_category'] == '<25']
gte_25_df = df_age_cat[df_age_cat['Age_category'] == '>=25']
```

```python
gte_25_df
```

```python
averages = lt_25_df.mean().to_dict()

# Add a new row with averages to the DataFrame
lt_25_df = lt_25_df.append(averages, ignore_index=True)
lt_25_df['Age_category'] = '<25'

averages = gte_25_df.mean().to_dict()

# Add a new row with averages to the DataFrame
gte_25_df = gte_25_df.append(averages, ignore_index=True)
gte_25_df['Age_category'] = '>=25'
```

```python
gte_25_df
```

```python
agecat_averages = pd.concat([lt_25_df.iloc[-1], gte_25_df.iloc[-1]], axis=1).transpose()
agecat_averages = agecat_averages.drop(columns=['email'])
```
